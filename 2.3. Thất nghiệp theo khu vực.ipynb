```python
import pandas as pd
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output

columns = ['Năm','Giới tính','Tỷ lệ']
df_region =pd.read_excel('Thất nghiệp 2011-2020.xlsx', 'Theo khu vực', header= None, names=columns)

columns_region = ['Năm','Nông thôn', 'Thành thị']
data_region =pd.read_excel('Thất nghiệp 2011-2020.xlsx', 'Lượng theo khu vực', header= None, names=columns_region)

app = dash.Dash(__name__)

options = [{'label': 'Tỷ lệ', 'value': 'Tỷ lệ'},
           {'label': 'Số lượng', 'value': 'Số lượng'}]

app.layout = html.Div([
    dcc.Dropdown(
        id='my-dropdown',
        options=options,
        value='Tỷ lệ'
    ),
    dcc.Graph(id='my-graph')
])

@app.callback(
    Output('my-graph', 'figure'),
    Input('my-dropdown', 'value')
)
def update_graph(value):
    if value == 'Tỷ lệ':
        fig = px.line(df_region, x='Năm', y='Tỷ lệ', color='Giới tính', title='Biểu đồ Tỷ lệ thất nghiệp theo khu vực của Việt Nam 2010-2021')
    else:
        fig =  px.bar(data_region, x='Năm', y=['Nông thôn', 'Thành thị'], title='Biểu đồ lượng người thất nghiệp theo khu vực của Việt Nam năm 2010-2021 (nghìn người)', barmode='stack')
        fig.update_layout(
    xaxis_title='Năm',
    yaxis_title='Số lượng (nghìn người)',
    legend_title_text='Khu vực',
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1
    ),
)
    return fig

if __name__ == '__main__':
    app.run_server(debug=False,port=8085)
```
